Importar pandas y altair

```python
import pandas as pd
import altair as alt
```

Obtener datos

```python
url ='https://docs.google.com/spreadsheets/d/e/2PACX-1vT06xF46sDjyjTfthKEhvz2PRjbNRIv_Nw2LB3qL01Bexqmeo09ErPN5eOkndWfUQ/pub?output=csv'
dataset = pd.read_csv(url)
dataset
```

```python
dataset.shape
```

```python
dataset.columns
```

```python
dataset = dataset.rename(columns = {'2007':'A2007'
                                    ,'2020':'A2020'
                              })
dataset
```

```python
alt.Chart(dataset).mark_rect().encode(
    y = "Departamento")
dataset[dataset['Departamento'] == 'Pasco']
```

```python
alt.Chart(dataset).mark_bar().encode(
    x ="Departamento:N" ,
    y = "A2007:Q"
    )
```

Titulo y color del grafico

```python
alt.Chart(dataset, height = 500, width = 800).mark_bar(color="firebrick").encode(
   alt.Y("A2007:Q", title='Salario Promedio 2007'),
    alt.X("Departamento:N", title='Departamentos')
)
   
```
